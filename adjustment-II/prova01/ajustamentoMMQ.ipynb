#### Voltando no raciocnio do :

- `1`  Tem mais de um meio de resolver MMQ, nesse caso explicado a seguir s  necessrio quando voc tem mais medidas que incgnitas, ou seja `N de Equaes < N de Incgnitas`.
- `2`  Como as medidas tem erro, e h mais medidas dq incgnitas, o sistema fica "sobredeterminado" e no existe uma soluo exata que satisfaa todas as medidas perfeitamente.
- `3`  O MMQ (VV = mnimo) encontra a soluo que mais se aproxima de todas as medidas, minimizando o total dos erros ao quadrado.

Frmula/Critrio matemtico para encontrar o valor mais provvel que melhor se ajusta a todos os dados, levando em conta a preciso de cada medio: <br>
 

| Varivel | Significado |
|---------|-------------|
| `V` | Vetor dos resduos (erros) |
| `V V` | Soma dos quadrados dos resduos |
| `P` | Matriz de pesos (confiana das medidas) |
| `V P V` | Soma dos quadrados dos resduos ponderada |

`V V = mnimo`  Minimiza a soma dos quadrados dos erros, `tratando todas as medidas como igualmente importantes`. <br>
`V P V = mnimo`  Minimiza a soma dos quadrados dos erros, mas `d mais importncia para as medidas mais precisas`.
