# Introdução ao Ajustamento pelo MMQ
Oberservações são ajustadas pois todas estão sujeitas a erros de medição.

`Erro/Resíduo (𝒗)  = Valor Medido (𝒍𝒏) – Valor Verdadeiro`

`Medida única ≠ Medida confiável`
O ajustamento das observações busca determinar um valor único e confiável para uma grandeza medida, a partir de dados imprecisos, além de estimar a precisão desse resultado.

| Variável | Significado |
|---------|-------------|
|  | Variáveis a serem estimadas* (ex: coordenadas, altitudes) |
|  | Precisão** das estimativas de `𝑋`  |
|  | Observações realizadas (ex: ângulos, distâncias, desníveis medidos) |
|  | Precisão das observações (ex: erro dos instrumentos) |

*`𝒍𝟏,𝒍𝟐,𝒍𝟑, … … … … … 𝒍𝒏` → valores obtidos de `𝒏` observações <br>
**ou incerteza (variância)

- Queremos os resultados, ou seja, os `𝑋`
- Cada observação `𝒍𝒏` tem um resíduo `𝒗_𝑖` (esses **𝒗_𝑖** são os **valores a priori desconhecidos**)
- Somando as observações com seus resíduos obtem-se o valor de `𝑋` (que são os resultados)

 ⭐

- É possível coletar observações/dados dos mesmos pontos por exemplo, quantas vezes quiser, cada vez que for feita uma nova observação, adiciona-se um `'` ao `X`, ou seja:
    - Conjunto de observações 1 → `X`
    - Conjunto de observações 2 → `X'`
    - Conjunto de observações 3 → `X'''`
    - Conjunto de observações 4 → `X''''` <br>
    Assim por diante...

- Como saber se uso `X', X'', X'''`??
    - Aceitar como melhor estimativa de `𝑿` o valor que torna mínimo a soma dos quadrados dos resíduos (GAUSS e LEGENDRE).
        - `TRADUZINDO DE MODO SIMPLES:` cada vez que faz `X', X''...` vai diminuindo os resíduos gradualmente, você para quando os resíduos não estiverem mais diminuindo tanto assim (a diferença vai ficando minúscula, ent n tem pq continuar se a melhora é ridiculamente pequena).

> `📎 FRASE MAIS IMPORTANTE DO MMQ:` No MMQ, queremos que a soma dos quadrados dos resíduos (erros) seja a menor possível.

#### Voltando no raciocínio do ⭐:

- `1º` → Tem mais de um meio de resolver MMQ, nesse caso explicado a seguir só é necessário quando você tem mais medidas que incógnitas, ou seja `Nº de Equações < Nº de Incógnitas`.
- `2º` → Como as medidas tem erro, e há mais medidas dq incógnitas, o sistema fica "sobredeterminado" e não existe uma solução exata que satisfaça todas as medidas perfeitamente.
- `3º` → O MMQ (VᵀV = mínimo) encontra a solução que mais se aproxima de todas as medidas, minimizando o total dos erros ao quadrado.

Fórmula/Critério matemático para encontrar o valor mais provável que melhor se ajusta a todos os dados, levando em conta a precisão de cada medição: <br>
 🌻

| Variável | Significado |
|---------|-------------|
| `V` | Vetor dos resíduos (erros) |
| `Vᵀ V` | Soma dos quadrados dos resíduos |
| `P` | Matriz de pesos (confiança das medidas) |
| `Vᵀ P V` | Soma dos quadrados dos resíduos ponderada |

🔹`Vᵀ V = mínimo` → Minimiza a soma dos quadrados dos erros, `tratando todas as medidas como igualmente importantes`. <br>
🔹`Vᵀ P V = mínimo` → Minimiza a soma dos quadrados dos erros, mas `dá mais importância para as medidas mais precisas`.

> `📎 OBSERVAÇÃO/CURIOSIDADE:` <br>
> O vetor dos resíduos, `V`, é assim: <br>
>  <br>
> Logo, usamos o `V` transposto, ou seja, `Vᵀ` para ficar assim: <br>
>  <br>
> Precisamos ele na forma vetor linha, ao invés de vetor coluna, pois precisamos fazer a soma do quadrado dos resíduos
>  <br>
> 
> Sendo que, precisamos fazer a soma do quadrado dos resíduos, pois:
> - `Elevar ao quadrado` para evitar que erros positivos e negativos se cancelem, dando uma visão real da magnitude dos erros.
> - `Somar` para combinar a contribuição de todos os resíduos em uma única medida do erro total, considerando o impacto conjunto de todas as medidas, e não apenas de uma por vez.
>
> E `P`, é uma matriz diagonal onde cada elemento da diagonal é o peso de uma medida. <br>
>  <br>
> Como o `P` usa os valores dos pesos `p1, p2 ... p𝒏`, utilizar a matriz `P` faz com que o cálculo que vc ta efetuando, dê mais importância às medidas mais precisas no ajuste
>

#### Voltando no raciocínio do 🌻:
Usamos o `Vᵀ V = mínimo` ou o `Vᵀ P V = mínimo` quando `Nº de Equações > Nº de Incógnitas`. PORÉM, quando `Nº de Equações = Nº de Incógnitas` OU `Nº de Equações > Nº de Incógnitas` temos que utilizar:

- `Nº de Equações > Nº de Incógnitas` <br>

    - Não tem solução exata → MMQ encontra a melhor solução aproximada
        - Fórmula do MMQ: <br>
        
        - > `📎 OBSERVAÇÃO/CURIOSIDADE: De onde vem essa fórmula acima?` <br>
        

- `Nº de Equações = Nº de Incógnitas` <br>

    - O que acontece → Temos solução única e exata
    - Quando usar →  Sistema determinado (sem redundância)
    - Não precisa do MMQ pois não há conflito entre medidas
